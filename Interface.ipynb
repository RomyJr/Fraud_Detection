```python
import sys
import pandas as pd
from PyQt5.QtWidgets import (
    QApplication, QMainWindow, QLabel, QLineEdit, QPushButton, QTextEdit, QMessageBox
)
from PyQt5.QtCore import Qt
from PyQt5.QtGui import QIcon

# Charger les données du fichier CSV
data_file = "C:/Users/user/Desktop/Scolarite/3 - E5e/S9/Projet_Fin_Etude/creditcard_2023.csv"
df = pd.read_csv(data_file)

# Classe principale de l'application
class FraudDetectionApp(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Détection de Fraude")
        self.setFixedSize(600, 800)  # Dimensions fixes de la fenêtre
        
        # Définir l'icône de la fenêtre
        self.setWindowIcon(QIcon("C:/Users/user/Desktop/Scolarite/3 - E5e/S9/Projet_Fin_Etude/Logo_PFE.png"))  # Remplacez par le chemin correct si nécessaire
        
        # Appliquer un style général
        self.setStyleSheet("""
            QMainWindow {
                background-color: #FF5050;  /* Couleur de fond principale */
            }
            QLabel {
                color: white;
                font-size: 14px;
                font-weight: bold;
            }
            QLineEdit {
                padding: 5px;
                font-size: 14px;
                border: 2px solid white;
                border-radius: 5px;
            }
            QPushButton {
                background-color: white;
                color: #FF5050;
                font-size: 14px;
                font-weight: bold;
                border: none;
                border-radius: 5px;
                padding: 5px;
            }
            QPushButton:hover {
                background-color: #FFC1C1;
            }
            QTextEdit {
                background-color: white;
                color: black;
                font-size: 14px;
                border-radius: 5px;
                padding: 10px;
            }
        """)

        # Widgets d'entrée et de contrôle
        self.label_id = QLabel("Entrez l'ID de la transaction :", self)
        self.label_id.setGeometry(20, 20, 250, 30)
        
        self.input_id = QLineEdit(self)
        self.input_id.setGeometry(270, 20, 200, 30)
        self.input_id.setPlaceholderText("Exemple : 0")
        
        self.button_check = QPushButton("Vérifier", self)
        self.button_check.setGeometry(480, 20, 100, 30)
        self.button_check.clicked.connect(self.display_transaction)
        
        # Zone d'affichage des détails de la transaction
        self.text_transaction = QTextEdit(self)
        self.text_transaction.setReadOnly(True)
        self.text_transaction.setGeometry(20, 70, 550, 575)  # Ajuster pour laisser de la place pour le résultat
        
        # Zone d'affichage du résultat (fraude ou non)
        self.label_result = QLabel("Résultat :", self)
        self.label_result.setAlignment(Qt.AlignLeft)
        self.label_result.setStyleSheet("""
            font-size: 30px; 
            font-weight: bold; 
            color: white;
        """)
        
        # Mettre à jour la position du label résultat
        self.update_result_position()

    def update_result_position(self):
        """Met à jour la position du label de résultat en fonction de la taille du QTextEdit."""
        text_edit_bottom = self.text_transaction.geometry().bottom() + 10  # 10 px de marge
        self.label_result.setGeometry(20, text_edit_bottom, 550, 30)

    def display_transaction(self):
        try:
            # Récupérer l'ID saisi
            transaction_id = int(self.input_id.text())
            if transaction_id < 0 or transaction_id >= len(df):
                raise ValueError("ID invalide. Veuillez entrer un ID valide.")
            
            # Obtenir les détails de la transaction
            transaction = df.iloc[transaction_id]
            transaction_details = "\n".join(f"{col}: {transaction[col]}" for col in df.columns)
            
            # Afficher les détails de la transaction
            self.text_transaction.setText(f"Transaction {transaction_id}:\n{transaction_details}")
            
            # Afficher un résultat statique pour l'instant
            self.label_result.setText("Résultat : Non frauduleuse")
        except ValueError as e:
            QMessageBox.critical(self, "Erreur", str(e))
        except Exception as e:
            QMessageBox.critical(self, "Erreur", f"Une erreur est survenue : {str(e)}")

# Exécution de l'application
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = FraudDetectionApp()
    window.show()
    sys.exit(app.exec_())
```
